```python
img = Image.open('input.png')
img = img.convert('RGB')
plt.imshow(img)
plt.show()
img = img.resize((28, 28))
np_img = np.array(img)
np_img = np.array([np_img])
np_img = np_img.astype('float32') / 255.0
```

```python
prediction = model.predict(np_img)
print(f'This is a {class_dict[np.argmax(prediction)]}')
```
