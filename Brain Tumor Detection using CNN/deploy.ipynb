```python
import numpy as np
from PIL import Image
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
deploy_model = tf.keras.models.load_model('model.h5')
```

```python
img = Image.open('input_image.jpg')
img = img.resize((150, 150))
img = img.convert('RGB')
np_img = np.array(img)
plt.imshow(np_img)
plt.show()
np_img = np.array([np_img])
np_img = np_img.astype('float32') / 255.0
```

```python
prediction = deploy_model.predict(np_img)
print(f'There is a {round(prediction[0][0]*100, 2)}% that there is a tumor')
```
